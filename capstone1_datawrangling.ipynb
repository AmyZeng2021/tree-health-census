#### This project uses data from the Food Environment Atlas. For more information, visit https://www.ers.usda.gov/foodatlas/

The data is comprehensive and covers the entire US. For this project specifically, the focus is on the state of *North Carolina* and the local foods grown and sold within the state.

```python
import pandas as pd
from pprint import pprint
```

```python
file = 'food-capstone-1.xls'
df = pd.ExcelFile(file) # loading the file
 
pprint(df.sheet_names)
```

Since we are looking into LOCAL produce in NORTH CAROLINA, we will only take data from the LOCAL sheet, in addition to 'Supplemental Data - County' and 'Supplemental Data - State'

```python
state_data = df.parse(3) # Supplemental Data - State

state_data[state_data['State'] == 'North Carolina']
```

```python
local_data = df.parse(-3) # LOCAL sheet

local_data[local_data['State'] == 'NC']
```
